```python
from langchain_core.prompts import PromptTemplate,FewShotPromptTemplate
```

```python
examples =[
    {
        "input" : "what is python ?",
        "output" : "python is the programming language. üêç"
    },
    {
        "input" : "what is javascript ?",
        "output" : "javascript is the web programming language. üåê"
    }
]
```

```python
template = PromptTemplate(
    input_variables=["input","output"],
    template="Q :{input} \nA :{output}"
)
```

```python
fewshot_template = FewShotPromptTemplate(
    examples = examples,
    example_prompt = template,
    prefix = "Answer the question with emoji at thr end",
    suffix = "Q: {input}\nA:",
    input_variables = ["input"]
)
fewshot_template
```

```python
prompt = fewshot_template.format(input = "what is ruby ?")
print(prompt)
```

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
llm = ChatGroq(
    model = "openai/gpt-oss-120b",
    api_key=os.getenv("GROK_API_KEY"),
)
```

```python
chain = fewshot_template | llm
```

```python
response = chain.invoke("what is ruby ?")
print(response.content)
```
